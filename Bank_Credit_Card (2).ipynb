##### Drop some Columns which are of no use :

#### Extract Independent and Dependent Variables :

```python
df.shape
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:, 23]
```

```python
x.head()
```

```python
y.head()
```

#### Spliting the dataset into the Training set and Test set :

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25 , random_state = 0)
```

### Step 2) Fitting Decision Tree Classifier to the Training Set :

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

### Step 3) Prediction of the test set result :  

```python
y_pred = dt.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

### Step 3) Creating Confusion Matrix : 

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

### Step 5) Accuracy :

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

### Step 6) Classification Report :


```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

### Step 7) Fitting Random Forest Classifier to the Training Set :

```python
from sklearn.ensemble import RandomForestClassifier
cr = RandomForestClassifier()
cr.fit(x_train, y_train)
```

### Step 8) Prediction of the test set result :

```python
y_pred = cr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

### Step 9) Creating Confusion Matrix : 

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

### Step 10) Accuracy :

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

### Step 11) Classification Report :

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
